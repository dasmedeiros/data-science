# 1. Load libraries and data


```python
#Load Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Data
df = pd.read_excel('./Hitters-Cleaned.xlsx', sheet_name='Sheet1')
df.head()
```

# 2. Key statistics

```python
#Key statistics for the dependent variable (Salary) grouped by League and Division
df.groupby(['League', 'Division'])['Salary'].describe()
```

```python
#Key statistics for the whole dataset
df.describe()
```

```python
#Key statistics transposed for better view
df.describe().T
```

# 3. Correlation 

```python
#Create and plot a correlation table
corr_table = df.corr()
corr_table
```

```python
#Plot a correlation heatmap
sns.heatmap(corr_table, xticklabels=corr_table.columns, yticklabels=corr_table.columns)
```

## 3.1 Correlation by league/division

```python
#Correlation between Salary and independent variables: American League East
df.loc[(df['League'] == 'A') & (df['Division'] == 'E')].corr().iloc[:-1, -1]
```

```python
#Correlation between Salary and independent variables: American League West
df.loc[(df['League'] == 'A') & (df['Division'] == 'W')].corr().iloc[:-1,-1]
```

```python
#Correlation between Salary and independent variables: National League East
df.loc[(df['League'] == 'N') & (df['Division'] == 'E')].corr().iloc[:-1,-1]
```

```python
#Correlation between Salary and independent variables: National League West
df.loc[(df['League'] == 'N') & (df['Division'] == 'W')].corr().iloc[:-1,-1]
```
